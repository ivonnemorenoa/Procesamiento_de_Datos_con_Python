### Casting

- Aprender a usar `astype`
- Aprender a lidiar con errores usando `to_numeric`
- Aprender a convertir `strings` e `ints` a `datetime`

```python
import pandas as pd
```

```python
df = pd.read_csv('../7. Transformación, filtración y ordenamiento de Datos/bestsellers.csv', index_col=0)

df.head()
```

```python
df.shape
# obtener la cantidad de filas y columnas de un DataFrame
```

```python
df['price'] = df['price'].map(str)
```

```python
df.dtypes
```

Podemos usar el método astype para pasarle a nuestro DataFrame un diccionario de conversión. Por ejemplo, vamos a convertir nuestras dos columnas de rank usando un diccionario de conversión. El tipo de dato que usamos para manejar weeks_on_list es el llamado datetime. Este tipo de dato nos permite manipular muy eficientemente.

```python
print(df.columns)
```

```python
diccionario_casting = {
    'rank': int,
    'weeks_on_list': int,
    'price': float
}
```

```python
df = df.astype(diccionario_casting)
```

```python
df
```
